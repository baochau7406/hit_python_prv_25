```python
# câu 8:  xếp chồng 2 mảng theo chiều dọc

import numpy as np

# tạo data 2 mảng 2d (2 hàng 5 cột)
# tạo mảng hai chiều (2 hàng 5 cột) từ 0 đến 9
arr1 = np.arange(10).reshape(2,-1)
print('arr1: \n', arr1)
# tạo mảng hai chiều (2 hàng 5 cột) chỉ có giá trị 1
arr2 = np.full((2,5),1)
print('arr2: \n',arr2)

# Cách 1 :
c =np.vstack((arr1,arr2))
print("Cách 1: \n", c)

# Cách 2 :
d =np.concatenate((arr1,arr2), axis=0)
print("Cách 2: \n", d)

# Cách 3 :
e =np.r_[arr1,arr2]
print("Cách 3 \n",e)

```

```python
# câu 15a : Tạo hàm xử lý trên mảng numpy

import numpy as np

def maxx(x,y):
    # Lấy số lớn nhất nhất trong 2 số
    return max(x, y)

a = np.array([5,7,9,8,6,4,5])
b = np.array([6,3,4,8,9,7,1])

# So sánh các cặp phần tử trong mảng a và b (theo index) sau đó lấy ra giá lớn hơn trong từng cặp
# dùng numpy.vectorize lấy hàm maxx: x,y và biến nó thành maxx: a[] -> b[].
pair_max = np.vectorize(maxx)
print(pair_max(a,b))
```

```python
# câu 22: Lấy vị trí của n giá trị đứng đầu từ một mảng numpy
import numpy as np

# cho mảng a chứa các số ngẫu nhiên từ [1, 50)
a = np.random.randint(1, 50, (5, 3))
print("Mảng a: \n", a)

# Flatten mảng để dễ lấy vị trí
flat = a.ravel()   # hoặc a.flatten()

# -------------------------
# Cách 1: Dùng np.argsort()
# -------------------------
print("\nCách 1 (argsort)")
idx = np.argsort(flat)[-5:]        # lấy 5 index lớn nhất
print("Vị trí :", idx)
print("Giá trị :", flat[idx])


# -------------------------
# Cách 2: Dùng np.argpartition(), giới thiệu thêm cho các bạn 1 cách khác ngoài argsort
# -------------------------
print("\nCách 2 (argpartition)")
idx2 = np.argpartition(flat, -5)[-5:]  # lấy 5 vị trí lớn nhất
print("Vị trí :", idx2)
print("Giá trị :", flat[idx2])

# Lưu ý kết quả là ngẫu nhiên nên kết quả có thể thay đổi mỗi khi chạy
```

```python
# câu 23: Chuyển đổi một mảng chứa nhiều mảng thành một mảng 1d phẳng

import numpy as np


# cho array_of_arrays như sau
arr1 = np.arange(3)
arr2 = np.arange(3,7)
arr3 = np.arange(7,10)

array_of_arrays = np.array([arr1, arr2, arr3], dtype=object)
print('array_of_arrays: ', array_of_arrays)

# Chuyển đổi array_of_arrays thành mảng 1d tuyến tính phẳng
#Cách 1: flatten
arr_2d = array_of_arrays.flatten()
print(arr_2d)
# #Cách 2 : ravel
arr_2d = array_of_arrays.ravel()
print(arr_2d)
# #Cách 3: reshape
arr_2d = array_of_arrays.reshape(-1)
print(arr_2d)
```

```python
import numpy as np

# Ma trận A và B
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[2, 0],
              [1, 2]])

# Tính AB và BA
AB = A@B
BA = A@A

print("A =\n", A)
print("B =\n", B)
print("AB =\n", AB)
print("BA =\n", BA)

print("AB == BA ? -->",np.array_equal(AB, BA) )
```

```python
import numpy as np

# Tạo hai ma trận A và B tùy ý
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8, 9],
              [1, 2, 3]])

print("A =\n", A)
print("B =\n", B)

# Tính (A + B)^T
left =(A + B).T

# Tính A^T + B^T
right =A.T + B.T

print("\n(A + B)^T =\n", left)
print("\nA^T + B^T =\n", right)

print("\nHai ma trận bằng nhau? ->",np.array_equal(left, right) )
```

```python
import numpy as np

# Ma trận A kích thước 2×3
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Ma trận B kích thước 3×2
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print("A =\n", A)
print("B =\n", B)

# Tính AB
AB =A@B
print("\nAB =\n", AB)

# Tính (AB)^T
left =AB.T
print("\n(AB)^T =\n", left)

# Tính B^T A^T
right =B.T @A.T
print("\nB^T A^T =\n", right )

print("\nHai kết quả bằng nhau? ->",np.array_equal(left, right))

```
